### Loop over countries for simplification

```python
import geopandas as gpd
import topojson as tp
from datetime import datetime
import numpy as np
```

```python
gdf = gpd.read_file("zip:///Users/tamarahuete/Documents/HALF_EARTH/WDPA_FILTERED_20210615_FILTERED_nomarine_with_oecm_wdpa.zip")
```

```python
gdf.dropna(subset = ["geometry"], inplace=True) #(217486-57)
```

```python
len(gdf)
```

```python
### Loop over countries to make the simplifications
len(gdf.ISO3.unique())
```

```python
%%time
simple = gpd.GeoDataFrame()
for country in gdf.ISO3.unique():
    start = datetime.now()
    print(f'starting {country} at {start.strftime("%H:%M:%S")}')
    test = gdf[gdf['ISO3'].isin([country])]
    print(f'Processing {len(test)} WPDAs')
    if len(test)> 2000:
        nwdpas= list(range(0,len(test)))
        splits= len(np.arange(0, len(test), 2000).tolist())
        for i in range(0,splits):
            first = np.array_split(nwdpas,splits)[i][0]
            last = np.array_split(nwdpas,splits)[i][-1]
            subtest = test[first:last]
            topo = tp.Topology(subtest, prequantize=False)
            gdf_simplified = topo.toposimplify(0.05).to_gdf()
            simple = simple.append(gdf_simplified)
            end = datetime.now()
            print(f'finished {country} split #{i} at {end.strftime("%H:%M:%S")}')
    else:
        if country == 'USA':
            nwdpas= list(range(0,len(test)))
            splits= len(np.arange(0, len(test), 500).tolist())
            for i in range(0,splits):
                first = np.array_split(nwdpas,splits)[i][0]
                last = np.array_split(nwdpas,splits)[i][-1]
                subtest = test[first:last]
                topo = tp.Topology(subtest, prequantize=False)
                gdf_simplified = topo.toposimplify(0.05).to_gdf()
                simple = simple.append(gdf_simplified)
                end = datetime.now()
                print(f'finished {country} split #{i} at {end.strftime("%H:%M:%S")}')
        else:     
            topo = tp.Topology(test, prequantize=False)
            gdf_simplified = topo.toposimplify(0.05).to_gdf()
            simple = simple.append(gdf_simplified)
            end = datetime.now()
            print(f'finished {country} at {end.strftime("%H:%M:%S")}')
    simple.to_file("/Users/tamarahuete/Documents/HALF_EARTH/wdpa_simple.shp")
```
